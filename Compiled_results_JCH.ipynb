```python
import numpy as np
import One_Source_vectorized_parallel as one
import Two_Sources_vectorized_parallel as two
import Two_Sources_vectorized_parallel_acceptor_source as acceptor
import Two_Sources_vectorized_parallel_vertical as vertical

```

## Single source case

```python
r = 1
alpha_l = 2
alpha_t = 0.05
beta = 1/(2*alpha_l)
C0 = 5
Ca = 8
gamma = 3.5

d = np.sqrt((r*np.sqrt(alpha_l/alpha_t))**2-r**2)
q = (d**2*beta**2)/4
# print(d,q)
n = 7             #Number of terms in mathieu series -1
M = 100

x_min = 0
x_max = 150
y_min = -10
y_max = 10
inc = 0.5 #resolution of the grid
```

```python
one.run(r_param=r,
        alpha_l_param=alpha_l,
        alpha_t_param=alpha_t,
        C0_param=C0,
        Ca_param=Ca,
        gamma_param=gamma,
        n_param=n, M_param=M,
        x_min = x_min,
        x_max = x_max,
        y_min = y_min,
        y_max = y_max,
        inc = inc)
```

## Two donor elements

```python
r = 1.0
alpha_l = 2.0
alpha_t = 0.05
beta = 1.0 / (2.0 * alpha_l)
C0 = 10.0
C1 = 10.0
Ca = 8.0
gamma = 3.5

d = np.sqrt((r * np.sqrt(alpha_l / alpha_t)) ** 2 - r ** 2)
q = (d ** 2 * beta ** 2) / 4.0

n = 7  # number of terms
M = 100  # boundary control points

# distance/offset for second source
x = 50.0
y = 10.0

#Domain Parameters
x_min = 0.0
x_max = 1200.0
y_min = -10.0
y_max = 20.0
inc = 0.5
```

```python
two.run(r_param=r,
        alpha_l_param=alpha_l,
        alpha_t_param=alpha_t,
        C0_param=C0,
        C1_param=C1,
        Ca_param=Ca,
        gamma_param=gamma,
        n_param=n,
        M_param=M,
        D1_param=x,
        D2_param=y,
        x_min=x_min, x_max=x_max, y_min=y_min, y_max=y_max, inc=inc
)
```

## Donor and acceptor element

```python
r = 1.0
alpha_l = 2.0
alpha_t = 0.05
beta = 1.0 / (2.0 * alpha_l)
C0 = 10.0
C1 = -8.0   #negative concentration for acceptor
Ca = 8.0
gamma = 3.5

d = np.sqrt((r*np.sqrt(alpha_l/alpha_t))**2 - r**2)
q = (d**2 * beta**2) / 4.0

n = 7
M = 100

#Element Position
x = 75.0
y = 0.0

#Domain Parameters
x_min = 0.0
x_max = 100.0
y_min = -5.0
y_max = 5.0
inc = 0.5
```

```python
acceptor.run(r_param=r,
    alpha_l_param=alpha_l,
    alpha_t_param=alpha_t,
    C0_param=C0,
    C1_param=C1,
    Ca_param=Ca,
    gamma_param=gamma,
    n_param=n,
    M_param=M,
    D1_param=x,
    D2_param=y,
    x_min=x_min, x_max=x_max, y_min=y_min, y_max=y_max, inc=inc
)
```

## Vertical model

```python
r = 1.0
alpha_l = 2.0
alpha_t = 0.005
beta = 1.0 / (2.0 * alpha_l)
C0 = 10.0
C1 = -C0
Ca = 8.0
gamma = 3.5

d = np.sqrt((r*np.sqrt(alpha_l/alpha_t))**2 - r**2)
q = (d**2 * beta**2) / 4.0

n = 9
M = 100

# vertical offset for second source; default set in run() to 2*r+0.2
x = 0.0
y = None  # will be set in run()

#Domain Parameters
x_min = 0.0
x_max = 600.0
y_min = -5.0
y_max = None
inc = 0.5
```

```python
vertical.run(r_param=r,
    alpha_l_param=alpha_l,
    alpha_t_param=alpha_t,
    C0_param=C0,
    C1_param=C1,
    Ca_param=Ca,
    gamma_param=gamma,
    n_param=n,
    M_param=M,
    D1_param=x,
    D2_param=y,         # default becomes 2*r + 0.2 if None
    x_min=x_min, x_max=x_max, y_min=y_min, y_max=y_max, inc=inc)
```
